```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests
from bs4 import BeautifulSoup
```

```python
docs=[]
for i in range(100):
    filename="emailscandals/doc"+str(i+1)+".txt"
    f = open(filename, 'rU')
    docs.append(f.read())
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
```

```python
tokenizer = RegexpTokenizer(r'\w+')

# create English stop words list
en_stop = get_stop_words('en')

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
```

```python
doc_set = docs
```

```python
tokenize = lambda doc: doc.lower().split(" ")
from sklearn.feature_extraction.text import TfidfVectorizer

sklearn_tfidf = TfidfVectorizer(ngram_range=(1, 2),norm='l2',min_df=0, use_idf=True, smooth_idf=False, sublinear_tf=True, tokenizer=tokenize)

sklearn_representation = sklearn_tfidf.fit_transform(doc_set)
#print tfidf_representation[0]
print sklearn_representation.toarray()[0].tolist()
#print document_0
```

```python
feature_array = np.array(sklearn_tfidf.get_feature_names())
tfidf_sorting = np.argsort(sklearn_representation.toarray()).flatten()[::-1]

n = 100
top_n = feature_array[tfidf_sorting][:n]
print top_n
```
